```python
import os
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image

# Define image size and other parameters
image_size = (224, 224)

# Image data augmentation
train_datagen = ImageDataGenerator(
    rescale=1. / 255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

# Specify the folder path
folder_path = 'processing-picture'  # Replace with the actual folder path

# Get all file names in the folder
image_files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]

# Iterate through all image files
for file_name in image_files:
    # Read the original image
    img_path = os.path.join(folder_path, file_name)
    original_img = image.load_img(img_path)  # Read the original image without resizing

    # Display the original image
    plt.figure()
    plt.imshow(original_img)
    plt.title(f"{file_name} (Original)")
    plt.axis('off')

    # Read and resize the image for preprocessing
    img = image.load_img(img_path, target_size=image_size)
    x = image.img_to_array(img)
    x = x.reshape((1,) + x.shape)

    # Apply image preprocessing and display
    i = 0
    for batch in train_datagen.flow(x, batch_size=1):
        print("Image Preprocessing:")
        print(" ") # This adds a blank line for better readability
        print("- Pixel Value Scaling: Scale pixel values to a range between 0 and 1 (dividing by 255).")
        print("- Random Rotation: Randomly rotate the image with a maximum rotation angle of 20 degrees.")
        print("- Horizontal Shift: Randomly shift the image horizontally with a maximum shift distance of 20% of the image width.")
        print("- Vertical Shift: Randomly shift the image vertically with a maximum shift distance of 20% of the image height.")
        print("- Shear Transformation: Apply a shear transformation to the image with a maximum shear angle of 20 degrees.")
        print("- Random Zoom: Randomly zoom the image with a maximum zoom factor of 20%.")
        print("- Horizontal Flip: Randomly flip the image horizontally with a probability of 50%.")
        plt.figure()
        plt.imshow(image.array_to_img(batch[0]))
        plt.title(f"{file_name}(After_Process)")
        plt.axis('off')
        i += 1
        if i % 1 == 0:
            break
    plt.show()
```
